### Lemmatization - It is similar technique like stemming. The output we will get after lemmatizaton is called 'lemma' which is a root word rather than the root stem.


### Nltk Provides WordNetLemmatizer class which is a thin wrapper around the wordnet corpus. This class uses morphy() function to the WordNet corpus Reader class to find a lemma. 

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
""" Here we have to use the Parts of speech(Pos)-
                        noun - n
                        adjective - a
                        verb - v
                        adverb - r """
lemma_word = lemmatizer.lemmatize("Clickable", pos = 'v') 
print(lemma_word)
```
