```python
## Stemming- Is the process of reducing a word to its word stem that affixes to suffixes or roots of the word known as lemma. 
# Stemming is the natural language understanding (NLU) and natural language processing(NLP).
```

```python
# Example: eating ---> eat
# Going ---> go
```

```python
words1 = ['stemming', 'Running', 'Numbers', 'Tokenization']
words1
```

```python
from nltk.stem import PorterStemmer
stemming = PorterStemmer()

for word in words1: 
    print(word+"--->"+stemming.stem(word))
```
