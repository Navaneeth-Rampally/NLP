### Tokenization: The process of breaking down complex/large paragraphs or sentences into simpler units called tokens.

### Topics covered:
1. Converting corpus into documents
2. Converting corpus into words.
3. converting documents into words
4. vocabulary --> Unique words



```python
Paragraph = """ Hello, Iam Navaneeth. Currently iam learning NLP concepts theoretically and practically.
Iam learning form krish naik sir, who is an instructor at ineuron platform. """
```

```python
print(Paragraph)
```

```python
#Tokenization
## corpus --> documents
from nltk.tokenize import sent_tokenize
documents = sent_tokenize(Paragraph)
documents
```

```python
## Paragraphs--> words
from nltk.tokenize import word_tokenize
words = word_tokenize(Paragraph)
words
```

```python
for sentence in documents:
    print(sent_tokenize(sentence))
```

```python
## documents --> words
for words in documents:
    print(word_tokenize(words))
```
