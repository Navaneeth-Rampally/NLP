Named entity recognition: is a NLP technique that automatically identifies and classifies named entities in text into predefined categories.


```python
import nltk
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
```

```python
sentence = "Iam Navaneeth Rampally. I was born on 11th july 2002."
 
```

```python
#step1: converting sentence into words
from nltk import word_tokenize, ne_chunk, pos_tag
words = word_tokenize(sentence)

#step2: performing Pos on words(!! Required step to perform NER )
poss_tag = pos_tag(words)

#step3: Now performing NER
ner_tree = ne_chunk(poss_tag)

#step4: printing the new_NER
print(ner_tree)
```
