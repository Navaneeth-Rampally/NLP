### Stemming: The process of converting any word into its word stem (Root word). 

### Topics covered: 
1. Porterstemmer --> It is a basic method of stemming.
2. RegexpStemmer --> With the help of this, we can convert any regular expression into its root word.
3. SnowBallStemmer --> This is a very accurate method of stemming compared to porterstemmer. 


```python
# Example: eating ---> eat
# Going ---> go
# programming--->program
```

```python
from nltk.stem import PorterStemmer

stemming = PorterStemmer()
words1 = ["program", 'Singer', "programming", 'Dancer', "programmers", "programs", "Drinking"]

for word in words1:
    print(word + "--->" + stemming.stem(word))
```

```python
stemming.stem("sitting")
```

```python
stemming.stem("standing")
```

```python
from nltk.stem import RegexpStemmer
regexp_stem = RegexpStemmer('ing$|s$|e$|able$', min = 3)
regexp_stem.stem('Eating')
```

```python
for word in words1:
    print(word+"--->"+regexp_stem.stem(word))
```

### Snowball Stemmer --> Similar to porterstemmer but it is more accurate.

```python
from nltk.stem import SnowballStemmer
snow = SnowballStemmer('english') ## we have to mention the language as a parameter 
snow.stem('Annoying')
```

```python
# Checking out the difference between PorterStemmer and SnowBallStemmer
# 1. Using porterstemmer
stemming.stem('fairly')
```

```python
# 2. Using SnowballStemmer()
snow.stem('Fairly')
```
